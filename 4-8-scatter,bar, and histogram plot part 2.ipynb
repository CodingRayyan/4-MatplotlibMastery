###
## Color and Transparency
Alpha refers to transparency (range should be from 0 to 1)

###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Rank']
plt.figure(figsize = (10,3))
a,plt.hist(a['College'], bins = 50, color = 'y', alpha = 0.895) 
```

###
## Boxplot
###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Rank']
plt.figure(figsize = (10,3))
plt.boxplot(a['Rank'])
plt.show()
```

###
## Filling Boxplot
###

```python

import matplotlib.pyplot as plt
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
b = a['College']
c = a['Rank']
plt.figure(figsize = (10,3))
plt.boxplot(a['Rank'], patch_artist = True)
plt.show()
```

###
## Boxprops
###

```python

import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')

plt.figure(figsize = (5,3))
box_properties = dict( color = 'c', linewidth = 10, linestyle = ':')
plt.boxplot( a['Rank'], patch_artist = True, boxprops = box_properties )
plt.show()
```
